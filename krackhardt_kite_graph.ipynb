# Krackhardt Kite Graph
Krackhardt Kite graph is a simple graph with ten nodes. It was named after David Krackhardt who was a researcher of social network theory.

It was introduced in 1990 to differeniate between different centrality concepts. The vertex with the the maximum **degree** is in the center (here labeled Dazzle, with 6 degrees), the vertex with maxium **betweenness centrality** (here labeled Harmony) and the two vertices with maximum **closeness centraility** (here labeled Frost and Glacier) all have different characteristics.

*Referenced from [Wikipedia](https://en.wikipedia.org/wiki/Krackhardt_kite_graph)*


```python
# import packages
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# plt size
plt.figure(figsize=(11,7))
# create graph object
G = nx.Graph()

# create edges
G.add_edge('Alicorn', 'Bolt')
G.add_edge('Alicorn', 'Crimson')
G.add_edge('Alicorn', 'Dazzle')
G.add_edge('Bolt', 'Alicorn')
G.add_edge('Bolt', 'Dazzle')
G.add_edge('Bolt', 'Ember')
G.add_edge('Crimson', 'Alicorn')
G.add_edge('Crimson', 'Dazzle')
G.add_edge('Crimson', 'Frost')
G.add_edge('Frost', 'Glacier')
G.add_edge('Frost', 'Dazzle')
G.add_edge('Frost', 'Crimson')
G.add_edge('Glacier', 'Ember')
G.add_edge('Glacier', 'Dazzle')
G.add_edge('Glacier', 'Frost')
G.add_edge('Glacier', 'Harmony')
G.add_edge('Ember', 'Bolt')
G.add_edge('Ember', 'Dazzle')
G.add_edge('Ember', 'Glacier')
G.add_edge('Harmony', 'Frost')
G.add_edge('Harmony', 'Glacier')
G.add_edge('Harmony', 'Ignatius')
G.add_edge('Ignatius', 'Harmony')
G.add_edge('Ignatius', 'Jasmine')
G.add_edge('Jasmine', 'Ignatius')

# Set node positions explicitly
pos = {
    'Alicorn': (0, 1),
    'Bolt': (0, -1),
    'Crimson': (1, 2),
    'Dazzle': (1, 0),
    'Ember': (1, -2),
    'Frost': (2, 1),
    'Glacier': (2, -1),
    'Harmony': (3, 0),
    'Ignatius': (4, 0),
    'Jasmine': (5, 0)
}

# nodes
nx.draw_networkx_nodes(G, 
                       pos, 
                       node_size=3000, 
                       node_color='#3B719F',
                       alpha=0.95,
                       node_shape='o')

# edges
nx.draw_networkx_edges(G, 
                       pos, 
                       alpha=0.5, 
                       width=4)

# labels
nx.draw_networkx_labels(G, 
                        pos, 
                        font_size=11, 
                        font_family='sans-serif',
                        font_color='w')
plt.title("Krackhardt Kite Graph", 
          fontdict={'fontsize':24})
plt.axis('off')
plt.savefig("./images/krackhardt_kite.png")
plt.show()  
```
